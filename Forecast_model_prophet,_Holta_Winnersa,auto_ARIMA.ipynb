```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
import numpy as np

# Загрузка данных
data = pd.read_excel('Задания_1_2.xlsx')
dates = pd.to_datetime(data['Date'])  # Преобразование столбца дат в формат datetime
series = data['series1']

# Подготовка данных для Prophet
data_prophet = pd.DataFrame({'ds': dates, 'y': series})

# Разделение данных на обучающую и тестовую выборки
train = data_prophet[data_prophet['ds'] <= '2017-06-30']
test = data_prophet[(data_prophet['ds'] > '2017-06-30') & (data_prophet['ds'] <= '2019-06-30')]

# Проверка размеров выборок
print(f"Размер обучающей выборки: {len(train)}")
print(f"Размер тестовой выборки: {len(test)}")

# Обучение модели на обучающей выборке
model = Prophet()
model.fit(train)

# Прогнозирование на тестовой выборке
future = model.make_future_dataframe(periods=len(test))
forecast = model.predict(future)

# Оценка качества прогноза на тестовой выборке
y_true = test['y'].values
y_pred = forecast['yhat'].values[-len(test):]

mae = mean_absolute_error(y_true, y_pred)
mape = mean_absolute_percentage_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print(f'MAE: {mae:.2f}')
print(f'MAPE: {mape:.2f}%')
print(f'RMSE: {rmse:.2f}')

# Прогнозирование на следующие 3 месяца (90 дней) после 30 июня 2019 года
future = model.make_future_dataframe(periods=180)
forecast = model.predict(future)

# Визуализация прогноза
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(train['ds'], train['y'], label='Train', color='blue')
ax.plot(test['ds'], test['y'], label='Test', color='orange')
ax.plot(forecast['ds'], forecast['yhat'], 'r--', label='Forecast')  # Прогноз красным цветом

# Форматирование оси X для отображения дат с интервалом в 7 дней
ax.xaxis.set_major_locator(mdates.DayLocator(interval=40))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()  # Автоматический поворот меток оси X

plt.title('Прогнозирование ежедневного поведения ряда с помощью Prophet')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error

# Загрузка данных
data = pd.read_excel('Задания_1_2.xlsx')
dates = pd.to_datetime(data['Date'])
series = data['series1']

# Подготовка данных
data_arima = pd.DataFrame({'ds': dates, 'y': series})
train = data_arima[data_arima['ds'] <= '2018-06-30']
test = data_arima[(data_arima['ds'] > '2018-06-30') & (data_arima['ds'] <= '2019-06-30')]

# Обучение модели ARIMA на обучающей выборке
model = ARIMA(train['y'], order=(2, 1, 2))  # Параметры (p, d, q) можно подбирать
model_fit = model.fit()

# Прогнозирование на тестовой выборке
forecast = model_fit.forecast(steps=len(test))
test['yhat'] = forecast

# Оценка качества прогноза на тестовой выборке
y_true = test['y'].values
y_pred = test['yhat'].values

mae = mean_absolute_error(y_true, y_pred)
mape = mean_absolute_percentage_error(y_true, y_pred)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))

print(f'MAE: {mae:.2f}')
print(f'MAPE: {mape:.2f}%')
print(f'RMSE: {rmse:.2f}')

# Прогнозирование на следующие 3 месяца после 30 июня 2019 года
future_dates = pd.date_range(start='2019-07-01', end='2019-09-30', freq='D')
future_forecast = model_fit.forecast(steps=len(future_dates))

# Визуализация прогноза
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(train['ds'], train['y'], label='Train', color='blue')
ax.plot(test['ds'], test['y'], label='Test', color='orange')
ax.plot(test['ds'], test['yhat'], 'r--', label='Forecast on Test')  # Прогноз на тестовой выборке красным цветом
ax.plot(future_dates, future_forecast, 'g--', label='Future Forecast')  # Прогноз на будущее зеленым цветом

# Форматирование оси X для отображения дат с интервалом в 30 дней
ax.xaxis.set_major_locator(mdates.DayLocator(interval=30))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()  # Автоматический поворот меток оси X

plt.title('Прогнозирование ежедневного поведения ряда с помощью ARIMA')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

# Загрузка данных
data = pd.read_excel('Задания_1_2.xlsx')
dates = pd.to_datetime(data['Date'])
series = data['series1']

# Разделение на обучающую и тестовую выборки
train_data = data[dates <= '2019-06-30']
test_data = data[dates > '2019-06-30']

# Построение модели экспоненциального сглаживания
model = ExponentialSmoothing(train_data['series1'], trend='add', seasonal='add', seasonal_periods=12)
model_fit = model.fit()

# Прогнозирование на следующие 30 дней после 30 июня 2019 года
future_dates = pd.date_range(start='2019-07-01', end='2019-07-30', freq='D')
future_forecast = model_fit.forecast(steps=len(future_dates))
future_forecast = pd.Series(future_forecast, index=future_dates)


# Визуализация прогноза
plt.figure(figsize=(12, 6))
plt.plot(train_data['Date'], train_data['series1'], label='Обучающая выборка')
plt.plot(future_dates, future_forecast, label='Прогноз на будущее', color='green')

# Форматирование оси X для отображения дат с интервалом в 30 дней
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=30))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()  # Автоматический поворот меток оси X

plt.legend()
plt.title('Прогнозирование с помощью модели экспоненциального сглаживания')
plt.grid(True)
plt.show()
```

2. Модель Хольта-Винтерса:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Загрузка данных
data = pd.read_excel('Задания_1_2.xlsx')
data['Date'] = pd.to_datetime(data['Date'])  # Убедитесь, что столбец Date имеет тип datetime
data['series1'] = pd.to_numeric(data['series1'], errors='coerce')  # Преобразование данных в числовой формат

# Разделение на обучающую и тестовую выборки
train_data = data[data['Date'] <= '2018-03-31']
test_data = data[data['Date'] > '2018-03-31']

# Построение модели Хольта-Винтерса
model = ExponentialSmoothing(train_data['series1'], trend='add', seasonal='add', seasonal_periods=122)
model_fit = model.fit()

# Прогнозирование
forecast = model_fit.forecast(steps=len(test_data))
forecast = pd.Series(forecast, index=test_data.index)

# Оценка качества прогноза
mse = mean_squared_error(test_data['series1'], forecast)
rmse = np.sqrt(mse)
mae = mean_absolute_error(test_data['series1'], forecast)
print(f'Mean Absolute Percentage Error: {mape:.2f}%')
print(f'Root Mean Squared Error: {rmse}')
print(f'Mean Absolute Error: {mae}')

# Визуализация прогноза
fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(train_data['Date'], train_data['series1'], label='Обучающая выборка', color='blue')
ax.plot(test_data['Date'], test_data['series1'], label='Тестовая выборка', color='orange')
ax.plot(test_data['Date'], forecast, 'r--', label='Прогноз на тестовой выборке')  # Прогноз на тестовой выборке красным цветом

# Форматирование оси X для отображения дат с интервалом в 30 дней
ax.xaxis.set_major_locator(mdates.DayLocator(interval=30))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()  # Автоматический поворот меток оси X

# Установка пределов для оси X
ax.set_xlim([pd.Timestamp('2015-01-01'), pd.Timestamp('2019-12-31')])

plt.legend()
plt.title('Прогнозирование с помощью модели Хольта-Винтерса')
plt.grid(True)
plt.show()
```
